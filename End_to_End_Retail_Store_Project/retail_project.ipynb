# Feature Engineering

```python
rt['order_year'] = rt.order_date.dt.year
rt['order_month'] = rt.order_date.dt.month
rt['order_month_name'] = rt.order_date.dt.month_name()
rt['order_day'] = rt.order_date.dt.day
rt['order_day_name'] = rt.order_date.dt.day_name()

rt['ship_year'] = rt.ship_date.dt.year
rt['ship_month'] = rt.ship_date.dt.month
rt['ship_month_name'] = rt.ship_date.dt.month_name()
rt['ship_day'] = rt.ship_date.dt.day
rt['ship_day_name'] = rt.ship_date.dt.day_name()
```

```python
rt.columns
```

```python
rt['quarter'] = rt.order_date.dt.quarter
```

```python
rt['profit_margin'] = rt.profit/rt.sales
```

```python
rt.columns
```

```python
pd.cut(rt.sales, bins=4).groupby(pd.cut(rt.sales, bins=4)).count()
```

```python
rt['order_value_category'] = pd.qcut(rt.sales, q=4)
```

```python
rt.columns
```

```python
rt['discount_level'] = pd.cut(rt.discount, bins=[0, .3, .65, 1], 
                                            labels=[1, 2, 3], include_lowest= True)
```

```python
rt[['discount', 'discount_level']].sort_values('discount', ascending=False).iloc[110:120, :3]
```
